```python
import requests
```

```python
doc_m='https://open-meteo.com/en/docs/cma-api#latitude=31.2222&longitude=121.4581&timezone=America%2FNew_York'

url='https://api.open-meteo.com/v1/forecast?latitude=52.52&longitude=13.41&current=temperature_2m,apparent_temperature&hourly=temperature_2m,apparent_temperature,weather_code'
```

```python
weather=requests.get(url)
```

```python
wd=weather.json()
wd
```

```python
print('This is in Shanghai')
```

```python
# it only has lattitudes and longitudes
```

```python
longitude=wd['longitude']
latitude=wd['latitude']
```

```python
print(f'This place is at {latitude} N {longitude} E.')
```

```python
current_t=wd['current']['temperature_2m']
current_f=wd['current']['apparent_temperature']
```

```python
difference=round(abs(current_t-current_f),2)
difference
```

```python
if current_t>current_f:
    print(f'It feels {difference} degrees colder.')
elif current_t<current_t:
    print(f'It feels {difference} degrees warmer.')
else:
    print('It feels exactly the same!')
```

```python
three_data=requests.get(three_url)
three=three_data.json()
```

```python
three
```

```python
three['daily']['time']
```

```python
maxi={}
date=three['daily']['time']
max_temp=three['daily']['temperature_2m_max']
```

```python
n=0
t={}
for day in date:
    t[max_temp[n]]=day
    n=n+1
t
```

```python
m=max(t.keys())
d=t[m]
```
